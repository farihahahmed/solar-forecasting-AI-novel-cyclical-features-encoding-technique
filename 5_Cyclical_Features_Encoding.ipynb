#Cyclical features encoding

```python
x
```

```python
from datetime import datetime

x['date'] = pd.to_datetime(x['date'], format='%d/%m/%y')
```

```python
x['month'] = x['date'].dt.month
```

```python
x['day'] = x['date'].dt.day
```

```python
x = x.drop(["date", "customer_num"], axis=1)
```

```python
y['date'] = pd.to_datetime(y['date'], format='%d/%m/%y')
```

```python
y['month'] = y['date'].dt.month
```

```python
y['day'] = y['date'].dt.day
```

```python
y = y.drop(["date", "customer_num"], axis=1)
```

#Converting to pi scale, then sin and cos

```python
import math
import numpy as np
x["month_norm"] = 2 * math.pi * x["month"] / x["month"].max()
x["month_cos"] = np.cos(x["month_norm"])
x["month_sin"] = np.sin(x["month_norm"])
```

```python
x["day_norm"] = 2 * math.pi * x["day"] / x["day"].max()
x["day_cos"] = np.cos(x["day_norm"])
x["day_sin"] = np.sin(x["day_norm"])
```

```python
x = x.drop(["month", "day"], axis=1)
```

```python
x = x.drop(["month_norm", "day_norm"], axis = 1)
```

```python
y["month_norm"] = 2 * math.pi * y["month"] / y["month"].max()
y["month_cos"] = np.cos(y["month_norm"])
y["month_sin"] = np.sin(y["month_norm"])
```

```python
y["day_norm"] = 2 * math.pi * y["day"] / y["day"].max()
y["day_cos"] = np.cos(y["day_norm"])
y["day_sin"] = np.sin(y["day_norm"])
```

```python
y = y.drop(["month", "day"], axis=1)
```

```python
y = y.drop(["month_norm", "day_norm"], axis = 1)
```

```python
y
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 1, shuffle = False)
```

#Normalization

```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(x_train)

# transform training data
x_train_norm = norm.transform(x_train)

# transform testing dataabs
x_test_norm = norm.transform(x_test)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np

#model
lr = LinearRegression()

# fit
lr.fit(x_train_norm, y_train)

# predict
pred = lr.predict(x_test_norm)

# RMSE
print("RMSE: " + str(np.sqrt(mean_squared_error(y_test,pred))))

#MAE
print("MAE: " + str(mean_absolute_error(y_test, pred)))
```

```python
#Finding pearson's correlation coefficient between
#real value and predicted value

y_test = np.array(y_test, dtype ='float32')
pred = np.array(pred, dtype ='float32')
new_y = y_test.reshape(565864, 1)
new_y = np.squeeze(new_y)
new_p = pred.reshape(565864, 1)
new_p = np.squeeze(new_p)

from scipy.stats import pearsonr
corr, _ = pearsonr(new_p, new_y)
print('Pearsons correlation: %.4f' % corr)
```

#KNN Model
